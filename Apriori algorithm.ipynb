```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('C:\\Users\\tomee\\Downloads\\datasets\\groceries_dataset.csv')
```

```python
grocery_df_date= df.groupby('Date').agg({'itemDescription':','.join})
```

```python
grocery_df_date
```

```python
grocery_df_date.head()
```

```python
#associative rule according to gouping by items sold on date
```

```python
grocery_df_date['itemDescription'] = grocery_df_date['itemDescription'].str.split(',')
```

```python
transactions=[]
for i in range(0,len(grocery_df_date)):
    transactions.append(grocery_df_date['itemDescription'][i])
```

```python
from apyori import apriori
rules = apriori(transactions=transactions,min_support=0.0042,min_confidence=0.8,min_lift=3,
                min_length=3,max_length=3)
```

```python
#visualizing results
```

```python
results = list(rules)
```

```python
results
```

```python
len(results)
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
